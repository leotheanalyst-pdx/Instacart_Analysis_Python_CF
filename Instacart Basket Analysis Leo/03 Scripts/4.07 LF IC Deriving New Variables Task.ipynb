## Using user-defined function on subset

```python
# Define a function

def price_label(row):
    if row['prices'] <=5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

## Use for-loop to create a column that summarizes how busy each day of the week is

```python
# Identify day of the week with most orders

ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

Value 0 means Saturday, therefore Saturday has highest number of orders and busiest day
Wednesday (value 4) has lowest frequency and is slowest day of the week

```python
# Create for-loop for busiest day

result = []

for value in ords_prods_merged['order_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

## Step 4 identify busiest hours and create new column

```python
# Identify busiest hours

ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

Observations:
Client wants three categories
Days 10, 11, 14, 15, 13, 12, 16, 9, and 17 all have above 200000 orders and will be grouped together as most
Days 8, 18, 19, 20, 7, 21, 22, 23, 6, 0, and 1 all have between 100000 and 200000 orders and will be grouped as average
Days 5, 2, 4, and 3 all have less than 100000 orders and will be grouped as fewest

Note, using actual number of orders rather than splitting the day equally into 3 categories by same number of hours do to the pattern in number of orders across hours of the day

```python
# create subset to run summary stats to check grouping choice

days_counts = ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
days_counts
```

```python
days_counts.describe()

# keeping choice to have 3 groups based on counts, but rather equal ranges of 0 to 100000, 100001 to 20000, and 200001 to 300000
```

```python
# Create for-loop for busiest hours

hours = []

for value in ords_prods_merged['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9, 17]:
        hours.append('Most orders')
    elif value in [5, 2, 4, 3]:
        hours.append('Fewest orders')
    else:
        hours.append('Average orders')
```

```python
hours
```

```python
# Merge hours list with dataframe as new column

ords_prods_merged['Busiest_period_of_day'] = hours
```
