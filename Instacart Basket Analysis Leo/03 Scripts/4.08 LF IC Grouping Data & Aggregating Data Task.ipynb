## Step 4 Create a loyalty flag for existing customers

```python
#Group data, transform values, and complete the new column for entire dataframe for customer order max values

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## Step 7 Create order frequency flag based on days_since_last_order

```python
# Reexamine columns for next steps

ords_prods_merge.head()
```

```python
# Transform data from days_since_last_order in new column

ords_prods_merge['median_last_order'] = ords_prods_merge.groupby(('user_id'))['days_since_last_order'].transform(np.median)
```

```python
# Check output

ords_prods_merge.head(50)
```

```python
# Create flag by days_since_last_order median

ords_prods_merge.loc[ords_prods_merge['median_last_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_last_order'] > 10) & (ords_prods_merge['median_last_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_last_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check value counts

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

Matches shape of dataframe, curious about NaN values

```python
ords_prods_merge[['days_since_last_order', 'median_last_order', 'frequency_flag']].head(50)
```

Note that NaN in days_since_last_order would mean ordered the same day with no prior orders. If that customer had orders after that, then they would have a numberical value to correspond to the frequency flag.

Should change NaN to not have a "frequent customer" value

Examine (5) NaN values from the frequency flag further

```python
print(ords_prods_merge[ords_prods_merge.frequency_flag.isnull()])
```

As expected, the null values come from a customer made their first and only order, therefore all NaN values in the median should be coded as "Non-frequent customers" in the frequency_flag

```python
# Change null values to 'Non-frequent customer'

ords_prods_merge['frequency_flag'].fillna('Non-frequent customer', inplace = True)
```

```python
# Check value counts

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
